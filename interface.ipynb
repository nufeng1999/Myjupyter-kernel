```python
##%overwritefile
##%file:src/ISpecialID.py
##%file:../jupyter-MyPython-kernel/jupyter_MyPython_kernel/ISpecialID.py
##%file:../jupyter-MyC-kernel/jupyter_MyC_kernel/ISpecialID.py
##%file:../jupyter-MyDart-kernel/jupyter_MyDart_kernel/ISpecialID.py
##%noruncode
from typing import Dict, Tuple, Sequence
from abc import ABCMeta, abstractmethod
class ITag(metaclass=ABCMeta):
    @abstractmethod
    def getName(self) -> str:
        pass
    @abstractmethod
    def getAuthor(self) -> str:
        pass
    @abstractmethod
    def getIntroduction(self) -> str:
        pass
    @abstractmethod
    def getPriority(self)->int:
        pass
    @abstractmethod
    def setKernelobj(self,kobj,magics):
        pass
##内核shutdown时调用
    @abstractmethod
    def on_shutdown(self, restart):
        pass
##编译代码时调用
    @abstractmethod
    def on_before_compile(self,code)->Tuple[bool,str]:
        pass
    @abstractmethod
    def on_after_compile(self,returncode,binfile)->bool:
        pass
##执行代码时调用
    @abstractmethod
    def on_before_exec(self,code)->Tuple[bool,str]:
        pass
    @abstractmethod
    def on_after_exec(self,returncode,execfile)->bool:
        pass



'''单标签接口'''
class IStag(ITag):
    @abstractmethod
    def getIDSptag(self) -> str:
        pass
    @abstractmethod
    def on_ISpCodescanning(self,key,line) -> str:
        pass
'''双标签接口'''
class IDtag(ITag):
    @abstractmethod
    def getIDDpbegintag(self) -> str:
        pass
    @abstractmethod
    def getIDDpendtag(self) -> str:
        pass
    ##在整理代码时调用双标签接口
    @abstractmethod
    def on_IDpReorgCode(self,line) -> str:
        pass

'''BOOL标签接口'''
class IBtag(ITag):
    @abstractmethod
    def getIDBptag(self) -> str:
        pass
    ##在扫描代码时调用
    @abstractmethod
    def on_IBpCodescanning(self,key,line) -> str:
        pass
```
