```python
##%overwritefile
##%file:src/ISpecialID.py
##%noruncode

from abc import ABCMeta, abstractmethod

class ISpecialID(metaclass=ABCMeta):
    @abstractmethod
    def getName(self) -> str:
        pass
    def getAuthor(self) -> str:
        pass
    @abstractmethod
    def getBrief(self) -> str:
        pass
    @abstractmethod
    def getPriority(self)->int:
        pass
    @abstractmethod
    def getIDSymbol(self) -> str:
        pass


    @abstractmethod
    def setKernelobj(self,obj):
        pass

    @abstractmethod
    def on_Codescanning(self,line) -> str:
        pass

    '''
        Return <0,Code execution was interrupted abnormally
        Return 0 ,If you continue to perform subsequent default operations normally
        Return 1 ,Do not proceed with subsequent operations
    '''
    @abstractmethod
    def on_Codeexecute(self,obj, code, silent, store_history=True,
                   user_expressions=None, allow_stdin=True)->int:
        pass
    @abstractmethod
    def on__shutdown(self, restart):
        pass




```
