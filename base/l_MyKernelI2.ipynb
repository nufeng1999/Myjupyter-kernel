```python
## IMyKernel
##%noruncode
##%overwritefile
## %file:../src/IMyKernel.py
## %file:../../jupyter-MyPython-kernel/jupyter_MyPython_kernel/IMyKernel.py
##%fnlist:CMake Mermaid
##%fnlist:Python C Dart Kotlin Groovy Java Batch PS R Swift Lua Perl PHP Tcl Vimscript M4 Dot
##%fnlist:Gjs Vala Nodejs Bash VBScript WLS Go Html TypeScript Make Ruby Rust
##%fileforlist:../../jupyter-My$fnlist-kernel/jupyter_My$fnlist_kernel/IMyKernel.py
##
##   MyPython Jupyter IMyKernel
## 
from typing import Dict, Tuple, Sequence,List, Set
from abc import ABCMeta, abstractmethod
class IMyKernel():
    ## 获得运行文件的类型 script,exe,class
    @abstractmethod
    def get_runfiletype(self)->str:
        pass
    ## 获得内核信息简称
    @abstractmethod
    def get_kernelinfo(self)->str:
        pass
    ## 获得语言信息字典对象
    @abstractmethod
    def get_language_info(self):
        pass
    
    ## 获得main()函数的头部
    @abstractmethod
    def get_main_head(self)->str:
        pass
    ## 获得main()函数的尾部
    @abstractmethod
    def get_main_foot(self)->str:
        pass

    ## 获得magics对象
    @abstractmethod
    def get_mymagics(self)->object:
        pass
    ## 设置magics对象
    @abstractmethod
    def set_mymagics(self,object):
        pass
    ## 执行次数
    @abstractmethod
    def get_execution_count(self):
        pass

    ## 信息输入接口
    @abstractmethod
    def rawinput(self):
        pass

    ## 发送到前端的应答
    @abstractmethod
    def sendresponse(self,contents,name='stdout',mimetype=None):
        pass

    ## 核心调用时的接口，执行代码时将回调jkobj里的do_execute函数
    @abstractmethod
    def do_execute(self, code, silent, store_history=True,
                   user_expressions=None, allow_stdin=True)->Dict:
        pass

    ## 核心重起时的接口
    @abstractmethod
    def do_shutdown(self, restart):
        pass

    ## 执行代码的接口
    @abstractmethod
    def do_runcode(self,return_code,file_name,
                    magics,code, 
                    silent, store_history=True,
                    user_expressions=None, 
                    allow_stdin=True)->Dict:
        pass
    ## 编译接口
    @abstractmethod
    def do_compile_code(self,return_code,file_name,
                    magics,code, silent, store_history=True,
                    user_expressions=None, 
                    allow_stdin=True)->Tuple[bool,Dict,Dict, str,str,str]:
        pass
    ## 创建代码文件的接口
    @abstractmethod
    def do_create_codefile(self,magics,code, 
                    silent, store_history=True,
                    user_expressions=None, 
                    allow_stdin=True)->Tuple[bool,Dict,Dict, str,str,str]:
        pass
    ## 预执行接口
    @abstractmethod
    def do_preexecute(self,code,magics,
                silent, store_history=True,
                user_expressions=None, 
                allow_stdin=False)->Tuple[bool,Dict,Dict,str]:
        pass





```
